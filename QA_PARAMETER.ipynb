```python
import subprocess, sys

for pkg in ("torch","transformers","datasets","fuzzysearch","numpy","pandas","scikit-learn","matplotlib"):
    try:
        __import__(pkg)
    except ImportError:
        subprocess.check_call([sys.executable, "-m", "pip", "install", pkg])


import time
import random
import re
import subprocess
import sys

import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from fuzzysearch import find_near_matches
from datasets import Dataset
from transformers import (
    DistilBertTokenizerFast,
    DistilBertForQuestionAnswering,
    TrainingArguments,
    Trainer,
    pipeline
)
import matplotlib.pyplot as plt



# Text-noise and sample generation
def introduce_noise(text, typo_prob=0.05):
    chars, i = list(text), 0
    while i < len(chars):
        if random.random() < typo_prob:
            op = random.choice(['swap','delete','replace'])
            if op=='swap' and i+1<len(chars):
                chars[i],chars[i+1]=chars[i+1],chars[i]; i+=1
            elif op=='delete':
                chars.pop(i); i-=1
            else:
                chars[i]=random.choice('abcdefghijklmnopqrstuvwxyz')
        i+=1
    s = ''.join(chars)
    if random.random()<0.1:
        ins = random.choice([' like',' you know',' um',' uh'])
        pos = random.randint(0,len(s))
        s = s[:pos] + ins + s[pos:]
    return s

TEMPLATE_LIST = [
    "Set a carbon tax of ${tax} per ton in {year}.",
    "In {year}, impose a tax of ${tax}/t.",
    "Apply ${tax} carbon levy by {year}.",
    "By year {year}, tax CO2 at ${tax} per tonne.",
    "What about charging ${tax} tax in {year}?"
]

def generate_noisy_samples(n=5000, seed=42):
    random.seed(seed)
    rows = []
    for _ in range(n):
        tax  = random.randrange(0, 201)  # 0-200
        year = random.randrange(2020, 2100)   # 2020-2099
        template = random.choice(TEMPLATE_LIST)
        clean = template.format(tax=tax, year=year)
        noisy = introduce_noise(clean)
        rows.append({
            "query": noisy,
            "tax":   str(tax),
            "year":  str(year)
        })
    return pd.DataFrame(rows)


TAXES = [str(t) for t in range(0, 201, 1)]
YEARS = [str(y) for y in range(2020, 2100)]

def extract_cascade(text):
    txt = re.sub(r"\b(like|you know|um|uh)\b", " ", text, flags=re.IGNORECASE)
    m_y = re.search(r"\b(20[2-9]\d)\b", txt)

    year = m_y.group(1) if m_y else None
    m_t = re.search(r"\$\s*(\d{1,3})", txt)
    tax  = m_t.group(1) if m_t else None
    # fuzzy fallback
    if year is None:
        for y in YEARS:
            if find_near_matches(y, txt, max_l_dist=1):
                year = y; break
    if tax is None:
        for tval in TAXES:
            if find_near_matches(tval, txt, max_l_dist=2):
                tax = tval; break
    return tax, year

def make_qa_records(df):
    recs = []
    for _, r in df.iterrows():
        for key in ['tax','year']:
            txt = str(getattr(r, key))
            start = r.query.find(txt)
            if start >= 0:
                recs.append({
                    'context': r.query,
                    'question': f"What is the {key}?",
                    'answers': {'text':[txt], 'answer_start':[start]}
                })
    return recs

def prepare_features(examples):
    enc = tokenizer(
        examples['question'], examples['context'],
        truncation='only_second', max_length=128, padding='max_length',
        return_offsets_mapping=True
    )
    starts, ends = [], []
    for i, offsets in enumerate(enc.offset_mapping):
        ans = examples['answers'][i]
        sc = ans['answer_start'][0]
        ec = sc + len(ans['text'][0])
        seq_ids = enc.sequence_ids(i)
        c0 = seq_ids.index(1)
        c1 = len(seq_ids)-1 - seq_ids[::-1].index(1)
        s_idx, e_idx = c0, c0
        for idx in range(c0, c1+1):
            if offsets[idx][0] <= sc < offsets[idx][1]: s_idx = idx
            if offsets[idx][0] < ec <= offsets[idx][1]: e_idx = idx
        starts.append(s_idx); ends.append(e_idx)
    enc['start_positions'] = starts
    enc['end_positions']   = ends
    return enc

# Single experiment function
def run_experiment(n_samples, num_train_epochs, per_device_train_batch_size, learning_rate, seed=42):
    # Generate and split data
    df = generate_noisy_samples(n_samples, seed)
    df_train, df_val = train_test_split(df, test_size=0.3, random_state=seed)
    train_ds = Dataset.from_list(make_qa_records(df_train))
    val_ds   = Dataset.from_list(make_qa_records(df_val))

    # Load tokenizer and model
    global tokenizer
    tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
    model = DistilBertForQuestionAnswering.from_pretrained('distilbert-base-uncased').to(device)

    # Preprocess
    train_enc = train_ds.map(prepare_features, batched=True, remove_columns=train_ds.column_names)
    val_enc   = val_ds.map(prepare_features, batched=True, remove_columns=val_ds.column_names)

    # Train
    args = TrainingArguments(
        output_dir=f'out_{n_samples}_{num_train_epochs}_{per_device_train_batch_size}_{learning_rate}',
        num_train_epochs=num_train_epochs,
        per_device_train_batch_size=per_device_train_batch_size,
        learning_rate=learning_rate,
        logging_steps=50,
        report_to=['none'],
        no_cuda=not torch.cuda.is_available()
    )
    trainer = Trainer(
        model=model, args=args,
        train_dataset=train_enc, eval_dataset=val_enc,
        tokenizer=tokenizer
    )
    start = time.time()
    trainer.train()
    runtime = time.time() - start

    # Evaluate
    qa_pipe = pipeline('question-answering', model=model, tokenizer=tokenizer, device=device)
    acc = {'tax':0,'year':0}
    for r in df_val.itertuples():
        t_c, y_c = extract_cascade(r.query)
        t_out = t_c or qa_pipe(question="What is the tax?", context=r.query)['answer'].strip()
        y_out = y_c or qa_pipe(question="What is the year?", context=r.query)['answer'].strip()
        if t_out == r.tax:  acc['tax'] += 1
        if y_out == r.year: acc['year'] += 1
    n_val = len(df_val)
    return acc['tax']/n_val, acc['year']/n_val, runtime

# Grid search (one parameter at a time)
device = 0 if torch.cuda.is_available() else -1
baseline = {
    'n_samples': 5000,
    'num_train_epochs': 3,
    'per_device_train_batch_size': 8,
    'learning_rate': 3e-5
}
param_grid = {
    'n_samples': [5000, 10000, 15000, 2000],
    'num_train_epochs': [3, 5, 7, 10],
    'per_device_train_batch_size': [4, 8, 16, 32],
    'learning_rate': [1e-5, 3e-5, 5e-5, 1e-4]
}

results = []
for param, values in param_grid.items():
    for val in values:
        cfg = baseline.copy()
        cfg[param] = val
        print(f"Running {param}={val}")
        tax_acc, year_acc, runtime = run_experiment(
            cfg['n_samples'], cfg['num_train_epochs'],
            cfg['per_device_train_batch_size'], cfg['learning_rate']
        )
        results.append({
            'param': param,
            'value': val,
            'tax_acc': tax_acc,
            'year_acc': year_acc,
            'runtime': runtime
        })

# Aggregate and plot

df_res = pd.DataFrame(results)
for p in param_grid:
    df_p = df_res[df_res['param'] == p].sort_values('value')

    # Accuracy
    plt.figure()
    plt.plot(df_p['value'], df_p['tax_acc'], label='tax')
    plt.plot(df_p['value'], df_p['year_acc'], label='year')
    plt.xlabel(p)
    plt.ylabel('Accuracy')
    plt.title(f'Accuracy vs {p}')
    plt.legend()
    plt.show()

    # Runtime
    plt.figure()
    plt.plot(df_p['value'], df_p['runtime'])
    plt.xlabel(p)
    plt.ylabel('Runtime (s)')
    plt.title(f'Runtime vs {p}')
    plt.show()
```

The performance of QA under the optimal combination parameters

```python

import subprocess, sys
for pkg in ("torch","transformers","datasets","fuzzysearch","numpy","pandas","scikit-learn"):
    try:
        __import__(pkg)
    except ImportError:
        subprocess.check_call([sys.executable, "-m", "pip", "install", pkg])

# --- imports ---
import time, random, re
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from fuzzysearch import find_near_matches
from datasets import Dataset
from transformers import (
    DistilBertTokenizerFast, DistilBertForQuestionAnswering,
    TrainingArguments, Trainer, pipeline
)


device = 0 if torch.cuda.is_available() else -1


random.seed(42)
torch.manual_seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(42)


# Noisy Data generation

def introduce_noise(text, typo_prob=0.05):
    """Character-level typos + occasional fillers ('like', 'um', ...)."""
    chars, i = list(text), 0
    while i < len(chars):
        if random.random() < typo_prob:
            op = random.choice(['swap','delete','replace'])
            if op=='swap' and i+1<len(chars):
                chars[i],chars[i+1]=chars[i+1],chars[i]; i+=1
            elif op=='delete':
                chars.pop(i); i-=1
            else:
                chars[i]=random.choice('abcdefghijklmnopqrstuvwxyz')
        i+=1
    s = ''.join(chars)
    if random.random() < 0.1:
        ins = random.choice([' like',' you know',' um',' uh'])
        pos = random.randint(0, len(s))
        s = s[:pos] + ins + s[pos:]
    return s

TEMPLATE_LIST = [
    "Set a carbon tax of ${tax} per ton in {year}.",
    "In {year}, impose a tax of ${tax}/t.",
    "Apply ${tax} carbon levy by {year}.",
    "By year {year}, tax CO2 at ${tax} per tonne.",
    "What about charging ${tax} tax in {year}?"
]

def generate_noisy_samples(n=2000, seed=42):

    random.seed(seed)
    rows = []
    for _ in range(n):
        tax  = random.randrange(0, 201)
        year = random.randrange(2020, 2100)
        clean = random.choice(TEMPLATE_LIST).format(tax=tax, year=year)
        rows.append({"query": introduce_noise(clean), "tax": str(tax), "year": str(year)})
    return pd.DataFrame(rows)


# Cascade (regex + fuzzy) used before QA

TAXES = [str(t) for t in range(0, 201)]
YEARS = [str(y) for y in range(2020, 2100)]

def extract_cascade(text):
    #Stage-1: regex + fuzzy backoff for tax/year

    txt = re.sub(r"\b(like|you know|um|uh)\b", " ", text, flags=re.IGNORECASE)
    m_y = re.search(r"\b(20[2-9]\d)\b", txt)             # 2020..2099
    year = m_y.group(1) if m_y else None
    m_t = re.search(r"\$\s*(\d{1,3})", txt)
    tax  = m_t.group(1) if m_t else None
    if year is None:
        for y in YEARS:
            if find_near_matches(y, txt, max_l_dist=1): year = y; break
    if tax is None:
        for t in TAXES:
            if find_near_matches(t, txt, max_l_dist=2): tax = t; break
    return tax, year


# Build QA records

def make_qa_records(df):
    recs = []
    for _, r in df.iterrows():
        for key in ["tax","year"]:
            s = r.query.find(str(r[key]))
            if s >= 0:
                recs.append({
                    "context": r.query,
                    "question": f"What is the {key}?",
                    "answers": {"text":[str(r[key])], "answer_start":[s]}
                })
    return recs

# Token map helper (global)

def prepare_features(examples):
    enc = tokenizer(
        examples["question"], examples["context"],
        truncation="only_second", max_length=128, padding="max_length",
        return_offsets_mapping=True
    )
    starts, ends = [], []
    for i, offsets in enumerate(enc.offset_mapping):
        ans = examples["answers"][i]
        sc, ec = ans["answer_start"][0], ans["answer_start"][0] + len(ans["text"][0])
        seq_ids = enc.sequence_ids(i)
        c0 = seq_ids.index(1)
        c1 = len(seq_ids)-1 - seq_ids[::-1].index(1)
        s_idx, e_idx = c0, c0
        for idx in range(c0, c1+1):
            if offsets[idx][0] <= sc < offsets[idx][1]: s_idx = idx
            if offsets[idx][0] <  ec <= offsets[idx][1]: e_idx = idx
        starts.append(s_idx); ends.append(e_idx)
    enc["start_positions"] = starts
    enc["end_positions"]   = ends
    return enc


#  run with BEST hyperparameters

def run_best_config(seed=42):
    # Best hyperparameters
    BEST = dict(
        n_samples=2000,
        num_train_epochs=3,
        per_device_train_batch_size=32,
        learning_rate=5e-5
    )

    # 70/30 split after noise injection
    df = generate_noisy_samples(BEST["n_samples"], seed)
    df_train, df_val = train_test_split(df, test_size=0.3, random_state=seed)

    # Build QA datasets
    train_ds = Dataset.from_list(make_qa_records(df_train))
    val_ds   = Dataset.from_list(make_qa_records(df_val))

    # Tokenizer and model
    global tokenizer
    tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")
    model = DistilBertForQuestionAnswering.from_pretrained("distilbert-base-uncased")
    if device == 0:
        model = model.to("cuda")


    train_enc = train_ds.map(prepare_features, batched=True, remove_columns=train_ds.column_names)
    val_enc   = val_ds.map(prepare_features,   batched=True, remove_columns=val_ds.column_names)

    # TrainingArguments with BEST params
    args = TrainingArguments(
        output_dir="qa_best_cfg",
        num_train_epochs=BEST["num_train_epochs"],
        per_device_train_batch_size=BEST["per_device_train_batch_size"],
        learning_rate=BEST["learning_rate"],
        logging_steps=50,
        report_to=["none"],
        no_cuda=(device == -1)
    )

    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=train_enc,
        eval_dataset=val_enc,
        tokenizer=tokenizer
    )

    # Fine-tuning
    t0 = time.time()
    trainer.train()
    train_time = time.time() - t0

    # Evaluation
    qa_pipe = pipeline("question-answering", model=model, tokenizer=tokenizer, device=device)
    acc = {"tax": 0, "year": 0}
    for r in df_val.itertuples():
        t_c, y_c = extract_cascade(r.query)
        t_out = t_c if t_c is not None else qa_pipe(question="What is the tax?",  context=r.query)["answer"].strip()
        y_out = y_c if y_c is not None else qa_pipe(question="What is the year?", context=r.query)["answer"].strip()
        if t_out == r.tax:  acc["tax"]  += 1
        if y_out == r.year: acc["year"] += 1
    n_val = len(df_val)
    return {"tax": acc["tax"]/n_val, "year": acc["year"]/n_val, "time_sec": train_time}


if __name__ == "__main__":
    metrics = run_best_config(seed=42)
    print("Best-config results (QA + cascade):", metrics)
```
